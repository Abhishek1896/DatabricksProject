#Define Ingestion Configuration


```python


INGESTION_CONFIG = [
    {
        "source": "crm",
        "path": "/Volumes/demo_project/bronze/raw_sources/cust_info.csv",
        "table": "crm_cust_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/demo_project/bronze/raw_sources/prd_info.csv",
        "table": "crm_prd_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/demo_project/bronze/raw_sources/sales_details.csv",
        "table": "crm_sales_details"
    },
    {
        "source": "erp",
        "path": "/Volumes/demo_project/bronze/raw_sources/CUST_AZ12.csv",
        "table": "erp_cust_az12"
    },
    {
        "source": "erp",
        "path": "/Volumes/demo_project/bronze/raw_sources/LOC_A101.csv",
        "table": "erp_loc_a101"
    },
    {
        "source": "erp",
        "path": "/Volumes/demo_project/bronze/raw_sources/PX_CAT_G1V2.csv",
        "table": "erp_px_cat_g1v2"
    }
]
```

#Creating Bronze Tables from Files

```python
for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} -> demo_project.{item['table']}")

    df = (
        spark.read
            .option("header","true")
            .option("inferSchema","true")
            .csv(item['path'])
    )

    (
        df.write
            .format("delta")
            .mode("overwrite")
            .saveAsTable(f"demo_project.bronze.{item['table']}")
    )
```
